```python
# data preprocessing
```

```python
df.shape
```

```python
data_training = pd.DataFrame(
    df['Close'][0 : int(len(df) * 0.7)]
)
#or data_training=pd.DataFrame(df.Close[0 : int(len(df) * 0.7)])
```

```python
data_testing = pd.DataFrame(
    df['Close'][int(len(df) * 0.7) : ]
)
```

```python
data_training
```

```python
data_testing
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
```

```python
data_training_array.shape
```

```python
#Data Sequence Creation
```

```python
x_train = []
y_train = []
for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i, 0])
```

```python
x_train, y_train = np.array(x_train), np.array(y_train)
```

```python
x_train.ndim
```

```python
y_train.ndim
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#Building the LSTM Model
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
from keras.models import Sequential
from keras.layers import Input, LSTM, Dense
```

```python
model = Sequential()
```

```python
model.add(Input(shape=(100, 1)))
```

```python
model.add(
    LSTM(
        units=128,
        activation='tanh',
        return_sequences=True
    )
)
```

```python
model.add(
    LSTM(units=64)
)
```

```python
model.add(Dense(25))
```

```python
model.add(Dense(1))
```

```python
# model training
```

```python
model.compile(
    optimizer='adam',
    loss='mean_squared_error'
)
```

```python
model.fit(
    x_train,
    y_train,
    epochs=50
)
```

```python
model.summary()
```

```python
model.summary()
```

```python
model.save('stock_prediction_model.keras') #saving the trained model
```

```python
# Preparing Test Data (Before Prediction)
```

```python
past_100_days = data_training.tail(100)
```

```python
final_df = pd.concat(
    [past_100_days, data_testing],
    ignore_index=True
)
```

```python
input_data = scaler.fit_transform(final_df)
```

```python
input_data.shape
# (856, 1)
```

```python
x_test = []
y_test = []
```

```python
for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i, 0])
```

```python
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
# making predictions:
```

```python
x_test[0].shape
# (100, 1)
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted
```

```python
y_predicted = scaler.inverse_transform(
    y_predicted.reshape(-1, 1)
).flatten()
```

```python
y_test = scaler.inverse_transform(
    y_test.reshape(-1, 1)
).flatten()
```

```python
#visualizing Plot Actual vs Predicted Prices
```

```python
plt.figure(figsize=(12,6))

plt.plot(y_test, color='blue', label='Original Price')
plt.plot(y_predicted, color='red', label='Predicted Price')

plt.xlabel('Days')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Model Evaluation
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_predicted)
print("Mean Squared Error (MSE):", mse)
```

```python
rmse = np.sqrt(mse)
print("Root Mean Squared Error (RMSE):", rmse)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_predicted)
print("R² Score:", r2)
```
